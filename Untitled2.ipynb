```python
# Exercício 1
numero = int(input("Digite um número de 4 algarismos: "))
milhar = numero // 1000
centena = (numero % 1000) // 100
dezena = (numero % 100) // 10
unidade = numero % 10
print("Milhar:", milhar)
print("Centena:", centena)
print("Dezena:", dezena)
print("Unidade:", unidade)

# Exercício 2
vel_max = float(input("Velocidade máxima (km/h): "))
velocidade = float(input("Velocidade do carro (km/h): "))
if velocidade <= vel_max:
    print("Dentro do limite.")
else:
    excesso = velocidade - vel_max
    print("Acima do limite por", excesso, "km/h.")

# Exercício 3
idade = int(input("Digite a idade: "))
if idade >= 18:
    print("Pode votar e dirigir.")
elif idade >= 16:
    print("Pode votar.")
else:
    print("Não pode votar nem dirigir.")

# Exercício 4
peso = float(input("Digite o peso dos peixes (em kg): "))
if peso > 50:
    excesso = peso - 50
    multa = excesso * 4
    print("Excesso:", excesso, "kg")
    print("Multa: R$", multa)
else:
    print("Peso dentro do limite.")

# Exercício 5
codigo = int(input("Código do operário: "))
horas = float(input("Horas trabalhadas: "))
if horas > 50:
    extras = horas - 50
    salario_excedente = extras * 20
    salario = (50 * 10) + salario_excedente
else:
    salario_excedente = 0
    salario = horas * 10
print("Salário excedente: R$", salario_excedente)
print("Salário total: R$", salario)

# Exercício 6
indice = float(input("Índice de poluição: "))
if indice >= 0.5:
    print("Todos os grupos devem parar.")
elif indice >= 0.4:
    print("Grupos 1 e 2 devem parar.")
elif indice >= 0.3:
    print("Grupo 1 deve parar.")
else:
    print("Índice aceitável.")

# Exercício 7
soma = 0
i = 1
while i <= 10:
    potencia = 2 ** i
    soma = soma + potencia
    i = i + 1
print("Soma:", soma)

# Exercício 8
n = int(input("Digite n (> 0): "))
soma = 0
i = 1
while i <= n:
    soma = soma + 1 / i
    i = i + 1
print("Soma:", soma)

# Exercício 9
i = 1000
while i >= 200:
    print(i)
    i = i - 3

# Exercício 10
maior = 0
menor = 10
i = 1
while i <= 50:
    nota = float(input("Nota " + str(i) + ": "))
    if 0 <= nota <= 10:
        if nota > maior:
            maior = nota
        if nota < menor:
            menor = nota
    i = i + 1
print("Maior nota:", maior)
print("Menor nota:", menor)

# Exercício 13
maior = 0
while True:
    idade = int(input("Digite uma idade (0 para parar): "))
    if idade == 0:
        break
    if idade > maior:
        maior = idade
print("Maior idade:", maior)

# Exercício 14
i = 0
while i < 1000:
    s = str(i)
    crescente = True
    j = 0
    while j < len(s) - 1:
        if s[j] >= s[j + 1]:
            crescente = False
        j = j + 1
    if crescente:
        print(i)
    i = i + 1

# Exercício 15
cont_m_s = 0
cont_f_s = 0
cont_m_n = 0
cont_f_n = 0
total = 0
while total < 2000:
    sexo = input("Sexo (m/f): ").lower()
    opniao = input("Gostou do produto? (s/n): ").lower()
    if sexo == "m" and opniao == "s":
        cont_m_s = cont_m_s + 1
    elif sexo == "f" and opniao == "s":
        cont_f_s = cont_f_s + 1
    elif sexo == "m" and opniao == "n":
        cont_m_n = cont_m_n + 1
    elif sexo == "f" and opniao == "n":
        cont_f_n = cont_f_n + 1
    total = total + 1
print("Homens que gostaram:", cont_m_s)
print("Mulheres que gostaram:", cont_f_s)
print("Homens que não gostaram:", cont_m_n)
print("Mulheres que não gostaram:", cont_f_n)
total_sim = cont_m_s + cont_f_s
total_nao = cont_m_n + cont_f_n
if total_sim > total_nao:
    print("Produto será lançado.")
else:
    print("Produto NÃO será lançado.")

# Exercício 16
c11 = 0
c12 = 0
c13 = 0
c14 = 0
c15 = 0
c16 = 0
nulo = 0
total = 0
while True:
    voto = int(input("Digite o número do candidato (0 para encerrar): "))
    if voto == 0:
        break
    elif voto == 11:
        c11 = c11 + 1
    elif voto == 12:
        c12 = c12 + 1
    elif voto == 13:
        c13 = c13 + 1
    elif voto == 14:
        c14 = c14 + 1
    elif voto == 15:
        c15 = c15 + 1
    elif voto == 16:
        c16 = c16 + 1
    else:
        nulo = nulo + 1
    total = total + 1
print("Candidato 11:", c11)
print("Candidato 12:", c12)
print("Candidato 13:", c13)
print("Candidato 14:", c14)
print("Candidato 15:", c15)
print("Candidato 16:", c16)
print("Votos nulos:", nulo)
maior = c11
ganhador = 11
if c12 > maior:
    maior = c12
    ganhador = 12
if c13 > maior:
    maior = c13
    ganhador = 13
if c14 > maior:
    maior = c14
    ganhador = 14
if c15 > maior:
    maior = c15
    ganhador = 15
if c16 > maior:
    maior = c16
    ganhador = 16
print("Candidato eleito:", ganhador)

# Exercício 17
x = float(input("Digite o valor de x: "))
y = int(input("Digite o valor de y: "))
resultado = 1
i = 0
if y >= 0:
    while i < y:
        resultado = resultado * x
        i = i + 1
else:
    while i < (-y):
        resultado = resultado * x
        i = i + 1
    resultado = 1 / resultado
print("Resultado:", resultado)

# Exercício 18
n = int(input("Digite quantos termos da sequência Fibonacci: "))
a = 1
b = 1
contador = 2
print(a)
print(b)
while contador < n:
    proximo = a + b
    print(proximo)
    a = b
    b = proximo
    contador = contador + 1

# Exercício 19
numero = 1000
while numero < 10000:
    parte1 = numero // 100
    parte2 = numero % 100
    soma = parte1 + parte2
    if soma * soma == numero:
        print(numero)
    numero = numero + 1

# Exercício 20
ano_atual = int(input("Digite o ano atual: "))
ano = 1
while ano <= ano_atual:
    if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
        print(ano)
    ano = ano + 1

# Exercício 21
base = int(input("Digite a base: "))
expoente = int(input("Digite o expoente (número inteiro): "))
resultado = 1
contador = 0
while contador < expoente:
    resultado = resultado * base
    contador = contador + 1
print("Resultado:", resultado)

# Exercício 22
soma = 0
quantidade = 0
maior = None
menor = None
while True:
    numero = float(input("Digite um número (0 para encerrar): "))
    if numero == 0:
        break
    soma = soma + numero
    quantidade = quantidade + 1
    if maior is None or numero > maior:
        maior = numero
    if menor is None or numero < menor:
        menor = numero
if quantidade > 0:
    media = soma / quantidade
    print("Média:", media)
    print("Maior:", maior)
    print("Menor:", menor)
else:
    print("Nenhum número foi digitado.")

# Exercício 23
i = 1
while i <= 10:
    print("Tabuada de", i)
    j = 1
    while j <= 10:
        print(i, "x", j, "=", i * j)
        j = j + 1
    print()
    i = i + 1
```
